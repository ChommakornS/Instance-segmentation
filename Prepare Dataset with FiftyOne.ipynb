```python
import fiftyone as fo
import fiftyone.zoo as foz
from fiftyone import ViewField as F
```

```python
# Load pig images and instance segmentation labels from open-image-v6
classes = [
    # YOUR CODE HERE - List a class name in string
    'Panda',
    'Dice',
    'Duck'
]
dataset = foz.load_zoo_dataset(
    "open-images-v7",
    split="validation",
    label_types=["segmentations"],
    classes=classes,
    max_samples=100,
    seed=42,
    shuffle=True,
    dataset_name="itcs498-oid-dataset",
)
```

```python
dataset
```

```python
# Explore your dataset and select/tag a subset for annotation
session = fo.launch_app(dataset)
```

```python
# # Get IDs of all samples in the view
# all_ids = session.view.values("id")
# selected_view = dataset.select(all_ids)

selected_view = dataset.select(session.selected)
selected_view
```

```python
# Create a view that only contains `classes`
filtered_view = selected_view.filter_labels(
    "ground_truth",
    F("label").is_in(classes))
```

```python
# Export dataset
import os

output_dir = '/content/drive/MyDrive/Colab Notebooks'  # TODO: Change this to be the path to your Google Drive
filtered_view.export(
    export_dir=os.path.join(output_dir, "data_l2"),
    dataset_type=fo.types.COCODetectionDataset,
    label_field="ground_truth",
    classes=classes,
    export_media=True,
)
```
